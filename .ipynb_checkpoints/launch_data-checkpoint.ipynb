```python
# loop through column
# to get rocket name, see below
# append to a list
# add list to df
df.rocket[0]['rocket_name']
df.launch_site[0]['site_name']
```

```python
rocket_df =pd.json_normalize(df.rocket)
# pd.json_normalize(df.telemetry)


rocket_df
```

```python
# loooop through this column
# becomes it's own df
# merge it to original



# loop through all the objects
rocket_name = []
nationality = []
site_name = []
site_id = []
payload_type = []

# insert loop here

# first loop, append all items in their respective lists
# 
# these loops grab all the data within each object
for column in df[['launch_site']]:
    columnSeriesObj = df[column]
    #print('Column Name: ', column)
    #print('Column Contents: ', columnSeriesObj.values)
    for each in columnSeriesObj.values:
        site_name.append(each['site_name'])
        site_id.append(each['site_id'])
        
    

for column in rocket_df[['rocket_name']]:
    columnSeriesObj = rocket_df[column]
    #print('Column Name: ', column)
    #rocket_name.append(columnSeriesObj)
    for each in columnSeriesObj.values:
        rocket_name.append(each)
# #     rocket_name.append(df.rocket[row]['rocket_name'])
# #     nationality.append(rocket_df["second_stage.payloads"][row][0]["nationality"])
# #     launch_site.append(df.launch_site[row]['site_name'])
# #     payload_type.append(rocket_df["second_stage.payload"])

for column in rocket_df[['second_stage.payloads']]:
    columnSeriesObj = rocket_df[column]
    #print('Column Name: ', column)
    for each in columnSeriesObj.values:
        nationality.append(each[0]['nationality'])
        payload_type.append(each[0]['payload_type'])
    
```

```python
new_df['Site_ID'] = site_id
new_df['Site_Name'] = site_name
new_df['Rocket_Name'] = rocket_name
new_df['Nationality'] = nationality
new_df['Payload_type'] = payload_type
```

```python
new_df


    
new_df.to_csv("new_launch_data.csv", index = False)
```
